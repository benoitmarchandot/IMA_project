### Features

J'ai choisi la plupart des features en m'inspirant des articles de recherche mis à notre disposition. Ils sont les suivants :
* f0, f1, f2 : les intensités moyennes dans chaque noyau pour chaque canal (on procède ici à un recalage d'histogramme sur le premier, pour avoir des valeurs cohérentes entre elles.)
* f3, f4, f5 : les intensités moyennes dans chaque cytoplasme pour chaque canal 
* f6, f7, f8 : les contrastes moyens dans chaque noyau pour chaque canal
* f9, f10, f11 : les contrastes moyens dans chaque cytoplasme pour chaque canal
* f12, f13 : les aires des noyaux et cytoplasmes calculée relativement aux tailles des images
* f14 : le nombre de coins du contour du noyau détectés par le module de corner detection d'open cv
* f15, f16 : les distances minimale et maximale entre deux coins
* f17, f18 : les coordonnées x et y du centre du noyau (on rescale chaque cellule pour un soucis de cohérence entre les données)
* f19, f20, f21, f22, f23 : la même chose avec le cytoplasme
* f24, f25 : les diamètres maximum de chaque noyaux et de chaque cytoplasme.

Ces données géométriques et radiométriques sont plus ou moins pertinentes, comme nous le verrons par la suite.

Regardons ce que font les fonctions d'open cv concernant la détection de coins, ainsi que l'identification du cercle inscrits dans le noyau et le cytoplasme.

```python
##### Shi-Tomasi corner detection
import numpy as np
import cv2
from matplotlib import pyplot as plt

img2 = np.zeros_like(im)
img2[:,:,0] = im_nuc
img2[:,:,1] = im_nuc
img2[:,:,2] = im_nuc
img = img2
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray,25,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,255,-1)

plt.imshow(img),plt.show()
corners = corners.tolist()
for i in range(len(corners)):
    corners[i] = corners[i][0]
corners = 1/(im_cyt.shape[0]*im_cyt.shape[1])*np.array(corners, dtype=float)

kmeans = KMeans(init='random', n_clusters=1, max_iter=10000, n_init=100).fit(corners)
center = kmeans.cluster_centers_
plt.plot(center[0,0], center[0,1],'o', color='r')
plt.show()
```

```python
y_train = y_train.astype('float32').reshape((-1,1))
# Layer sizes
n_input = Xn_train.shape[1]  # Number of input features
n_hidden = 50  # Number of hidden nodes

# create the model 

model_mlp = Sequential() 
model_mlp.add(Dense(n_hidden, input_shape=(n_input, )))
model_mlp.add(Activation('relu'))
model_mlp.add(Dense(1, input_shape=(n_hidden,)))
model_mlp.add(Activation('sigmoid'))

# create the loss and optimiser
learning_rate = 0.01
model_mlp.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(lr=learning_rate),metrics=["accuracy"])

# Run optimisation algorithm
n_epochs = 20
batch_size = 64

print('Training')
model_mlp.fit(Xn_train, y_train, batch_size=batch_size, epochs=n_epochs) 
score = model_mlp.evaluate(Xn_test, y_test, verbose=False)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
